# Practical Session: Data Manipulation with pandas


---

## 1. Data Exploration Phase
*Goal: Understand the scale and health of your dataset.*

* **`df.head(n)`**: View the first `n` rows to verify data loading and column headers.
* **`df.info()`**: The "X-ray" of your data. Shows data types and identifies missing values.
* **`df.shape`**: A quick attribute check for the number of (Rows, Columns).
* **`df.describe()`**: Instant statistical summary (Mean, Min, Max) for all numeric columns.

---

## 2. Data Cleaning Phase
*Goal: Address missing or "null" values to prevent errors in analysis.*

* **`df.isnull().sum()`**: Map and count exactly how many missing values exist in each column.
* **`df.fillna(value)`**: Plug the holes. Replace `NaN` with a specific value (like 0 or the mean).
* **`df.dropna()`**: The "clean slate" approach. Remove any rows that contain missing data.

---

## 3. Transformation & Selection Phase
*Goal: Modify the data and extract specific insights.*

* **Creating Columns**: Learn how to create new data (e.g., `df['total'] = df['qty'] * df['price']`).
* **Boolean Masking**: Use logical conditions (e.g., `df[df['price'] > 100]`) to filter your data instantly.
